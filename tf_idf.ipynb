```python
import pandas as pd
from math import log
import os

path = '/Users/choiboyoung/소융캡/넷플릭스_리뷰_크롤링/단어불용어'

docs = []
f_name = []
for filename in os.listdir(path):
    with open(os.path.join(path, filename), 'r', encoding= 'utf-8') as f:
        docs.append(f.read())
    f_name.append(filename[:-6])
print(f_name)

vocab = list(set(w for doc in docs for w in doc.split()))
vocab.sort()
```

```python
N = len(docs) 

def tf(t, d):
  return d.count(t)

def idf(t):
  df = 0
  for doc in docs:
    df += t in doc
  return log(N/(df+1))

def tfidf(t, d):
  return tf(t,d)* idf(t)

print(N)
```

```python
result = []

for i in range(N):
  result.append([])
  d = docs[i]
  for j in range(len(vocab)):
    t = vocab[j]
    result[-1].append(tf(t, d))

tf_ = pd.DataFrame(result, columns = vocab)
tf_.index = f_name
print(tf_)
```

```python
result = []
for j in range(len(vocab)):
    t = vocab[j]
    result.append(idf(t))

idf_ = pd.DataFrame(result, index=vocab, columns=["IDF"])
print(idf_)
```

```python
result = []
for i in range(N):
  result.append([])
  d = docs[i]
  for j in range(len(vocab)):
    t = vocab[j]
    result[-1].append(tfidf(t,d))

tfidf_ = pd.DataFrame(result, columns = vocab)
tfidf_.index = f_name
tfidf_
```

```python
for i in range(N):
    with open ("/Users/choiboyoung/소융캡/넷플릭스_리뷰_크롤링/" + f_name[i] + "_tfidf", 'w', encoding='utf-8') as f:
        result = tfidf_.iloc[i, :].sort_values(ascending = False)
        result = result[:20]
        for j in result.index:
            print(j)
            f.write(j)
            f.write('\n')
    f.close()
```

```python
import pandas as pd
import xlsxwriter

writer = pd.ExcelWriter('final.xlsx', engine='xlsxwriter')

for i in range(N):
    df = pd.read_csv('/Users/choiboyoung/소융캡/넷플릭스_리뷰_크롤링/' + f_name[i] + '.csv')
    word_list = []
    
    with open('/Users/choiboyoung/소융캡/넷플릭스_리뷰_크롤링/' + f_name[i] + '_tfidf', 'r') as f:
        for line in f:
            word_list.append(line.strip())

    result = pd.DataFrame(columns=['word', 'score'])

    for j in range(len(df)):
        for word in word_list:
            if word in str(df['title'][j]) or word in str(df['content'][j]):
                result = result.append({'word': word, 'score': df['score'][j]}, ignore_index=True)
                
    result = result[result['score'] != '-']
    
    print(f_name[i])
    print(result)
    
    result['score'] = result['score'].astype(float)
    
    # groupby() 메소드로 그룹화
    grouped = result.groupby('word')

    # word 열에서 값들의 갯수와 그 값의 점수들을 합한 평균값 구하기
    mean_result = grouped.agg({'word': 'count', 'score': 'mean'})
    
    mean_result = mean_result.rename(columns={'word': 'word_count', 'score': 'score_mean'})
    mean_result = mean_result.reset_index()
    
    mean_result = mean_result.sort_values(by='word_count', ascending=False)
    
    print(mean_result)
        
    mean_result.to_excel(writer, sheet_name=f'sheet{i}', index=False)
writer.save()
```
